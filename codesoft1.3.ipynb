```python
class ATM:
    def __init__(self, balance=0):
        self.balance = balance

    def check_balance(self):
        print(f"\nüí∞ Current Balance: ‚Çπ{self.balance}")

    def deposit(self, amount):
        if amount > 0:
            self.balance += amount
            print(f"\n‚úÖ ‚Çπ{amount} deposited successfully.")
        else:
            print("\n‚ùå Invalid deposit amount.")

    def withdraw(self, amount):
        if amount <= 0:
            print("\n‚ùå Invalid withdrawal amount.")
        elif amount > self.balance:
            print("\n‚ö†Ô∏è Insufficient balance.")
        else:
            self.balance -= amount
            print(f"\n‚úÖ ‚Çπ{amount} withdrawn successfully.")


# Interactive ATM Interface
def atm_interface():
    atm = ATM(1000)  # Starting balance ‚Çπ1000

    while True:
        print("\n===== üè¶ ATM MENU =====")
        print("1. Withdraw Money")
        print("2. Deposit Money")
        print("3. Check Balance")
        print("4. Exit")
        choice = input("Select an option (1-4): ")

        if choice == "1":
            try:
                amount = float(input("Enter withdrawal amount: ‚Çπ"))
                atm.withdraw(amount)
            except ValueError:
                print("\n‚ùå Please enter a valid number.")

        elif choice == "2":
            try:
                amount = float(input("Enter deposit amount: ‚Çπ"))
                atm.deposit(amount)
            except ValueError:
                print("\n‚ùå Please enter a valid number.")

        elif choice == "3":
            atm.check_balance()

        elif choice == "4":
            print("\nüí≥ Thank you for using the ATM. Goodbye!")
            break

        else:
            print("\n‚ùå Invalid choice. Please select 1-4.")


# Run the ATM
atm_interface()
```

```python
class BankAccount:
    def __init__(self, account_holder, balance=0):
        self.account_holder = account_holder
        self.balance = balance

    def deposit(self, amount):
        if amount <= 0:
            print("\n‚ùå Deposit amount must be positive.")
        elif amount > 50000:
            print("\n‚ö†Ô∏è Deposit limit exceeded! Maximum ‚Çπ50,000 per transaction.")
        else:
            self.balance += amount
            print(f"\n‚úÖ ‚Çπ{amount} deposited successfully.")

    def withdraw(self, amount):
        if amount <= 0:
            print("\n‚ùå Withdrawal amount must be positive.")
        elif amount > 20000:
            print("\n‚ö†Ô∏è Withdrawal limit exceeded! Maximum ‚Çπ20,000 per transaction.")
        elif amount > self.balance:
            print("\n‚ö†Ô∏è Insufficient balance.")
        else:
            self.balance -= amount
            print(f"\n‚úÖ ‚Çπ{amount} withdrawn successfully.")

    def display_balance(self):
        print(f"\nüí∞ Current Balance: ‚Çπ{self.balance}")


class ATM:
    def __init__(self, bank_account):
        self.bank_account = bank_account

    def get_valid_amount(self, prompt):
        """Ensure user enters a valid positive numeric value."""
        while True:
            try:
                amount = float(input(prompt))
                if amount <= 0:
                    print("\n‚ùå Please enter a positive amount.")
                else:
                    return amount
            except ValueError:
                print("\n‚ùå Invalid input. Please enter a number.")

    def menu(self):
        while True:
            print("\n===== üè¶ ATM MENU =====")
            print("1. Withdraw Money")
            print("2. Deposit Money")
            print("3. Check Balance")
            print("4. Exit")
            choice = input("Select an option (1-4): ")

            if choice == "1":
                amount = self.get_valid_amount("Enter withdrawal amount: ‚Çπ")
                self.bank_account.withdraw(amount)

            elif choice == "2":
                amount = self.get_valid_amount("Enter deposit amount: ‚Çπ")
                self.bank_account.deposit(amount)

            elif choice == "3":
                self.bank_account.display_balance()

            elif choice == "4":
                print("\nüí≥ Thank you for using the ATM. Goodbye!")
                break

            else:
                print("\n‚ùå Invalid choice. Please select 1-4.")


# Example usage
user_account = BankAccount("Soumyadip Bhunia", 5000)
atm_machine = ATM(user_account)
atm_machine.menu()
```
