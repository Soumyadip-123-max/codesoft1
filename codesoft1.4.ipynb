```python
import json
import os

class Student:
    def __init__(self, name, roll_number, grade, age, department):
        self.name = name
        self.roll_number = roll_number
        self.grade = grade
        self.age = age
        self.department = department

    def display_info(self):
        print(f"Name: {self.name}")
        print(f"Roll Number: {self.roll_number}")
        print(f"Grade: {self.grade}")
        print(f"Age: {self.age}")
        print(f"Department: {self.department}")
        print("-" * 30)

    def to_dict(self):
        return {
            "name": self.name,
            "roll_number": self.roll_number,
            "grade": self.grade,
            "age": self.age,
            "department": self.department
        }

    @staticmethod
    def from_dict(data):
        return Student(data["name"], data["roll_number"], data["grade"], data["age"], data["department"])


class StudentManagementSystem:
    def __init__(self, filename="students.json"):
        self.students = []
        self.filename = filename
        self.load_data()

    def add_student(self, student):
        self.students.append(student)
        print(f"✅ Student {student.name} added successfully.")
        self.save_data()

    def edit_student(self, roll_number):
        for student in self.students:
            if student.roll_number == roll_number:
                print("Editing student information (press Enter to keep current value):")
                name = self.get_non_empty_input(f"Name [{student.name}]: ", allow_empty=True) or student.name
                grade = self.get_valid_grade(f"Grade [{student.grade}]: ", allow_empty=True) or student.grade
                age_input = input(f"Age [{student.age}]: ").strip()
                age = student.age
                if age_input:
                    if age_input.isdigit() and int(age_input) > 0:
                        age = int(age_input)
                    else:
                        print("⚠ Invalid age! Keeping old value.")
                department = self.get_non_empty_input(f"Department [{student.department}]: ", allow_empty=True) or student.department

                student.name = name
                student.grade = grade
                student.age = age
                student.department = department

                print("✅ Student information updated successfully.")
                self.save_data()
                return
        print("❌ Student not found.")

    def remove_student(self, roll_number):
        for student in self.students:
            if student.roll_number == roll_number:
                self.students.remove(student)
                print(f"✅ Student {student.name} removed successfully.")
                self.save_data()
                return
        print("❌ Student not found.")

    def search_student(self, roll_number):
        for student in self.students:
            if student.roll_number == roll_number:
                print("Student found:")
                student.display_info()
                return
        print("❌ Student not found.")

    def display_all_students(self):
        if not self.students:
            print("⚠ No students in the system.")
        else:
            print("📋 Student List:")
            for student in self.students:
                student.display_info()

    def save_data(self):
        data = [student.to_dict() for student in self.students]
        with open(self.filename, "w") as file:
            json.dump(data, file, indent=4)

    def load_data(self):
        if os.path.exists(self.filename):
            with open(self.filename, "r") as file:
                try:
                    data = json.load(file)
                    self.students = [Student.from_dict(d) for d in data]
                    print("📂 Data loaded successfully.")
                except json.JSONDecodeError:
                    print("⚠ Error reading data file. Starting fresh.")
        else:
            print("📂 No saved data found. Starting fresh.")

    # =========================
    # Validation Helper Methods
    # =========================
    def get_non_empty_input(self, prompt, allow_empty=False):
        while True:
            value = input(prompt).strip()
            if value or allow_empty:
                return value
            print("⚠ This field cannot be empty.")

    def get_valid_grade(self, prompt, allow_empty=False):
        valid_grades = {"A", "B", "C", "D", "F"}
        while True:
            grade = input(prompt).strip().upper()
            if allow_empty and not grade:
                return grade
            if grade in valid_grades:
                return grade
            print("⚠ Invalid grade! Please enter A, B, C, D, or F.")

    def is_unique_roll_number(self, roll_number):
        return all(student.roll_number != roll_number for student in self.students)


def main():
    sms = StudentManagementSystem()

    while True:
        print("\n===== Student Management System =====")
        print("1. Add Student")
        print("2. Edit Student")
        print("3. Search Student")
        print("4. Display All Students")
        print("5. Remove Student")
        print("6. Exit")
        choice = input("Enter your choice (1-6): ")

        if choice == "1":
            name = sms.get_non_empty_input("Enter name: ")
            while True:
                roll_number = sms.get_non_empty_input("Enter roll number: ")
                if sms.is_unique_roll_number(roll_number):
                    break
                print("⚠ Roll number already exists. Please enter a unique roll number.")
            grade = sms.get_valid_grade("Enter grade (A/B/C/D/F): ")
            while True:
                age_input = input("Enter age: ").strip()
                if age_input.isdigit() and int(age_input) > 0:
                    age = int(age_input)
                    break
                print("⚠ Age must be a positive integer.")
            department = sms.get_non_empty_input("Enter department: ")

            student = Student(name, roll_number, grade, age, department)
            sms.add_student(student)

        elif choice == "2":
            roll_number = input("Enter roll number of student to edit: ").strip()
            sms.edit_student(roll_number)

        elif choice == "3":
            roll_number = input("Enter roll number to search: ").strip()
            sms.search_student(roll_number)

        elif choice == "4":
            sms.display_all_students()

        elif choice == "5":
            roll_number = input("Enter roll number to remove: ").strip()
            sms.remove_student(roll_number)

        elif choice == "6":
            print("👋 Exiting Student Management System. Goodbye!")
            break

        else:
            print("⚠ Invalid choice! Please enter a number between 1 and 6.")


if __name__ == "__main__":
    main()
```
